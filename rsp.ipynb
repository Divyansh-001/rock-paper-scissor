```python
def game():
    from random import randint
    i=1
    player=0
    computer=0
    print("GAME STARTS")
    print("ENTER NUMBER OF ROUND")
    z=int(input())
    g=0
    h=0
    while i<=z:
        x=int(input())
        y=randint(1,3)
        match={1:"ROCK",2:"SCISSOR",3:"PAPER"}
        print(f'{match[x]} vs {match[y]}')
        player,computer=check(x,y)
        if player>computer:
            print("player wins")
            g=g+1
        elif computer>player:
            print("computer wins")
            h=h+1
        else:
            print("draw")
        i=i+1
    round_check(g,h)

def check(x, y):
    player=0
    computer=0
    if x==1 and y==1:
        player=1
        
        computer=1
    elif x==1 and y==2:
        player=1
    elif x==1 and y==3:
        computer=1
    elif x==2 and y==1:
        computer=1
    elif x==2 and y==2:
        player=1
        computer=1
    elif x==2 and y==3:
        player=1
    elif x==3 and y==1:
        player=1
    elif x==3 and y==2:
        computer=1
    elif x==3 and y==3:
        player=1
        computer=1
    return(player,computer)

def round_check(g,h):
    if g>h:
        print(" ")
        print("player wins the round")
    elif h>g:
        print(" ")
        print("computer wins the round")
    else:
        print(" ")
        print("draw")
    

def rules():
    print("RULES ARE SIMPLE")
    print("YOU ARE PLAYING WITH COMPUTER")
    print("INORDER TO WIN")
    print("YOU HAVE TO REMIND ")
    print("PAPER BEATS ROCK")
    print("SCISSORS BEATS PAPER")
    print("ROCK BEATS SCISSORS")
    print("PRESS 1. ROCK")
    print("PRESS 2. SCISSOR")
    print("PRESS 3. PAPER")
    print("1 ROUND CONTAINS 3 SETS")
    game()



x=1
print("WELCOME")
while x!=3:
    print("PRESS 1 FOR GAME")
    print("PRESS 2 FOR GAME RULES")
    print("PRESS 3 FOR EXIT")
    x=int(input("ENTER YOUR CHOICE "))
    if x==1:
        game()
    elif x==2:
        rules()
    
```
